```
#Import the dataset about the topics of Data Analysis
path = 'https://cocl.us/datascience_survey_data'

df = pd.read_csv(path, index_col=0)

if df.empty:
    print("The reading was unsuccessful")
else:
    print("Your dataframe is ready")
```

```
df.sort_values(by='Very interested', ascending=False, axis=0, inplace=True)
df
```

```
# The exercise is asking for the percenteges. So there were 2233 total participats, we count the % for all the rows: 
df['Very interested'] = round((df['Very interested'] / 2233 * 100), 2) 
df['Somewhat interested'] = round((df['Somewhat interested'] / 2233 * 100), 2) 
df['Not interested'] = round((df['Not interested'] / 2233 * 100), 2)
```

```
#Let's drop the toal for the graph
#df.drop(labels='Total',axis=1,inplace=True)
#df
```

```
ax = df.plot(kind='bar',
            figsize=(20, 8),
            width=0.8,
            color=['#5cb85c', '#5bc0de', '#d9534f'])

#The text manipulation: 
ax.set_title("Percenteges of Responders' Interests in Data Science Areas", fontsize=16) 
ax.set_yticks([])
ax.xaxis.set_tick_params(labelsize=14)
ax.legend(fontsize=14)


# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x(), i.get_height()+.5, \
            str(i.get_height())+'%', fontsize=14,
                color='dimgrey')

#Remove borders:
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
```
